```python
import pandas as pd
data=pd.read_csv("sonar1.csv",encoding="ISO-8859-1",names=[i for i in range(0,61)])  
```

```python
data.drop_duplicates(inplace=True)
data.isnull().sum()
```

```python
data.head()
```

```python
X=data.iloc[:,data.columns!=60].values
print(X)
```

```python

Y=data[60].replace(to_replace=["R","M"],value=[1,0])
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.neural_network import MLPClassifier
mlp=MLPClassifier(hidden_layer_sizes=(20,),alpha=1e-5,learning_rate_init=0.1,tol=1e-4,verbose=True,solver='sgd',activation='logistic',random_state=1)            
```

```python
mlp.fit(X_train,Y_train)
```

```python
pred=mlp.predict(X_train)
pred1=mlp.predict(X_test)
print(pred,"\n",pred1)
```

```python
print(Y_train,"\n",Y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score
print(accuracy_score(Y_train,pred)*100)
print("\n",confusion_matrix(Y_train,pred))
print("\n",f1_score(Y_train,pred)*100)
```

```python

print(accuracy_score(Y_test,pred1)*100)
print("\n",confusion_matrix(Y_test,pred1))
print("\n",f1_score(Y_test,pred1)*100)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,pred1))
```
